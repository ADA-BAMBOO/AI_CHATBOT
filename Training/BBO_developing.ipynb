```python
def readF(file_path):
  content=None
# -----------------docx------------------
  if file_path[-4:]=="docx":
    doc = Document(file_path)
    content = doc.paragraphs[0].text
    for paragraph in doc.paragraphs[1:]:
        if paragraph.text != '':
          content+="\n"+paragraph.text
    content=content.split("\n")
    for i in range(len(content)):
      content[i]=content[i].split("/:/")
    content = list(filter(lambda x: len(x)==2,content))
    content = np.matrix(content)
  # -----------------txt------------------
  if file_path[-3:]=="txt":
    with open(file_path,"r",encoding="utf-8")as file:
      content= file.read()
    content=content.split("\n")
  return content

def synthetic(path):
  text_path=path+"Text/"
  text_path_list =[text_path+i for i in os.listdir(text_path)]
  text_list=list(np.matrix(readF(i)) for i in text_path_list)

  link_path=path+"Link/"
  link_path_list =[link_path+i for i in os.listdir(link_path)]
  link_list=list(np.matrix(readF(i)) for i in link_path_list)
  return text_list,link_list
# -----------------txt------------------
  if file_path[-3:]=="txt":
    with open(file_path,"r",encoding="utf-8")as file:
      content= file.read()
    content=content.split("\n")
  return content
```

```python
def unique_matrix(mt):
  tmp=np.array(mt)
  unique_k1_values, indices_k1 = np.unique(tmp[:, 0], return_index=True)
  tmp=tmp[np.sort(indices_k1)]
  unique_k1_values, indices_k1 = np.unique(tmp[:, 1], return_index=True)
  mt=mt[np.sort(indices_k1)]
  return mt
```

```python
def Complete_Link(link_list):
  tmp=[]
  for i1,links in enumerate(link_list):
    tmp.append(links)
    for i2,link in enumerate(links):
      sub_link=docs_gov_Link_Extraction(link[0,1],i1,len(link_list),i2,len(links))
      tmp.append(unique_matrix(sub_link))
  return tmp
```

```python
def merge_short_sentences(sentences, threshold_min,threshold_max):
    merged_sentences = []
    i = 0

    while i < len(sentences):
        current_sentence = sentences[i]
        word_count = len(current_sentence.split())

        # Nếu là câu cuối cùng và nó quá ngắn, gộp vào câu trước đó
        if i == len(sentences) - 1 and word_count < threshold_min:
            if merged_sentences:
                merged_sentences[-1] += "." + current_sentence
            else:
                merged_sentences.append(current_sentence)  # Nếu chỉ có một câu, giữ nguyên
        elif word_count < threshold_min:
            if i + 1 < len(sentences):  # Kiểm tra xem có câu tiếp theo không
                sentences[i + 1] = current_sentence + "." + sentences[i + 1]
            else:
                merged_sentences.append(current_sentence+".")  # Không có câu tiếp theo thì vẫn giữ lại
        else:
            merged_sentences.append(current_sentence+".")

        i += 1  # Di chuyển sang câu tiếp theo

    return merged_sentences
```

```python
class Build_Data:
    waiting_list = []
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    model_que_name = "valhalla/t5-base-qg-hl"
    model_que = T5ForConditionalGeneration.from_pretrained(model_que_name).to(device)
    tokenizer_que = T5Tokenizer.from_pretrained(model_que_name)

    model_ans_name="deepset/roberta-large-squad2"
    model_ans = AutoModelForQuestionAnswering.from_pretrained(model_ans_name).to(device)
    tokenizer_ans = AutoTokenizer.from_pretrained(model_ans_name)

    model_para_name="facebook/bart-large-cnn"
    model_para = BartForConditionalGeneration.from_pretrained(model_para_name).to(device)
    tokenizer_para = BartTokenizer.from_pretrained(model_para_name)


    def __init__(self, Dic_contexts,epoch=1):  # Tăng phạm vi token để có câu dài hơn
        self.epoch=epoch
        self.contexts = list(Dic_contexts.values())[0]
        self.data=[]
        self.Context_To_Data()

    def Answering(self,context,question):
        nlp = pipeline('question-answering', model=self.model_ans, tokenizer=self.tokenizer_ans)
        context_chunks = self.chunk_text(context,self.tokenizer_ans,chunk_size=64)
        answers = []
        for chunk in context_chunks:
            answer = nlp({'context': chunk, 'question': question})
            answers.append(answer['answer'])

        final_answer = ' '.join(answers[1:])
        return self.paraphasing(final_answer)

    def chunk_text(self,text,tokenizer,chunk_size=512):
      tokens = tokenizer.encode(text)
      chunks = [tokens[i:i + chunk_size] for i in range(0, len(tokens), chunk_size)]
      return [tokenizer.decode(chunk) for chunk in chunks]

    def paraphasing(self,answer):
      nlp = pipeline('text2text-generation', model=self.model_para, tokenizer=self.tokenizer_para)

      # Generating a more coherent version
      output = nlp(f"paraphrase: {answer}")

      # Print the improved sentence
      return output[0]['generated_text']

    def Context_To_Data(self):
        for i in range(self.epoch):
          for index, context in enumerate(self.contexts):
            print(f"\rTiến độ sinh Data: {index+1}/{len(self.contexts)}", end="", flush=True)
            question=self.Questioning(context)
            answer=self.Answering(context,question)
            self.data.append({"context":context,"question":question,"answer":answer})


    def Questioning(self, context):
        input_text = f"generate question: {context}"
        input_ids = self.tokenizer_que.encode(input_text, return_tensors="pt").to(self.device)
        outputs = self.model_que.generate(
            input_ids,
            max_length=300,  # Tăng độ dài câu hỏi
            do_sample=True,
            top_k=40,
            top_p=0.95,
            temperature=0.6,
            repetition_penalty=1.6,
            num_return_sequences=1
        )
        question = self.tokenizer_que.decode(outputs[0], skip_special_tokens=True)
        return question


    def cal_token(self, text):
        return sum(len(sentence.split()) for sentence in text)
```

```python
import torch
from torch.utils.data import Dataset
import numpy as np

class ChatDataset(Dataset):
    def __init__(self, conversations, tokenizer, max_length=2048):
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.inputs = []
        self.raw_data = conversations
        # System prompt mặc định
        self.system_prompt = "You are a helpful, respectful and honest assistant. Always answer as helpfully as possible."
        for conv in conversations:
            # Validate dữ liệu
            if not self._is_valid(conv):
                print(f"❌ Bỏ qua dữ liệu không hợp lệ: {conv}")
                continue

            context = str(conv['context']).strip()
            question = str(conv['question']).strip()
            answer = str(conv['answer']).strip()

            # Format theo chuẩn LLaMA Instruct
            full_prompt = self._format_prompt(context, question, answer)
            # Tokenize toàn bộ prompt
            encoding = tokenizer(
                full_prompt,
                padding="max_length",
                truncation=True,
                max_length=self.max_length,
                return_tensors="pt"
            )
            input_ids = encoding['input_ids'].squeeze(0)
            attention_mask = encoding['attention_mask'].squeeze(0)

            # Tạo labels (mask phần prompt đầu)
            labels = input_ids.clone()

            # Tìm vị trí bắt đầu phần answer (sau <|start_header_id|>assistant<|end_header_id|>)

            prompt_until_answer = self._format_prompt(context, question, answer=None)
            answer_start = len(
                tokenizer(prompt_until_answer, truncation=True, max_length=self.max_length)['input_ids']
            )
            labels[:answer_start] = -100  # mask phần prompt, chỉ tính loss từ answer

            self.inputs.append({
                "input_ids": input_ids,
                "attention_mask": attention_mask,
                "labels": labels
            })

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):
        return self.inputs[idx]

    def get_question_answer(self, idx):
        if idx >= len(self.raw_data):
            return "", ""
        return self.raw_data[idx].get('question', ''), self.raw_data[idx].get('answer', '')

    def _is_valid(self, conv):
        if not isinstance(conv, dict):
            return False
        for key in ['context', 'question', 'answer']:
            if key not in conv:
                return False
            val = conv[key]
            if val is None or (isinstance(val, float) and np.isnan(val)):
                return False
            if str(val).strip().lower() == "nan":
                return False
        return True
    def _format_prompt(self, context, question, answer=None):
        """
        Format chuẩn theo Meta LLaMA-3-Instruct (chatml-style)
        """
        prompt = (
          "<|begin_of_text|><|start_header_id|>system<|end_header_id|>\n"
          f"{self.system_prompt}<|eot_id|><|start_header_id|>user<|end_header_id|>\n"
          f"CONTEXT:{context}\n\nQUESTION:{question}<|eot_id|><|start_header_id|>assistant<|end_header_id|>\n"
        )
        if answer:
          if answer.strip().endswith("<|end|>"):
            answer = answer.strip()[:-len("<|end|>")].strip()
          prompt += f"{answer}<|eot_id|>"
        return prompt
```

```python
class BBO_Training:
  save_directory = "/content/drive/MyDrive/HuyAnh/Model_Finetuning_RENEW_2"
  report_path="/content/drive/MyDrive/HuyAnh/Report/report.xlsx"
  rag_off=True


  def __init__(self
               ,data_list,Chat=False
               ,model_info=None
               ,tokenizer=None
               ,renew=False
               ,checkpoint_dir="/content"
               ,system_prompt="You are an expert assistant on the Cardano blockchain. When available, use the provided context to answer the user's question as accurately and clearly as possible. If the context is not enough or missing, use your own knowledge to give the best possible answer. Always respond helpfully and honestly."
               ,rag_dict=None
               ,rag_embedder=None):

    if rag_dict:
      self.rag_dict=rag_dict
      self.rag_embedder=rag_embedder
      self.rag_off=False

    if Chat:
      self.system_prompt = system_prompt
      self.rag_dict=data_list
      self.model_id = model_info
      self.rag_embedder = SentenceTransformer("mixedbread-ai/mxbai-embed-large-v1")
      self.tokenizer = AutoTokenizer.from_pretrained(self.model_id)
      self.tokenizer.pad_token = self.tokenizer.eos_token
      self.model = AutoModelForCausalLM.from_pretrained(
          self.model_id,
          torch_dtype=torch.bfloat16,
          device_map="auto"
      )
      return

    else:
      self.checkpoint_dir=checkpoint_dir
      self.system_prompt = system_prompt
      self.data_list=data_list
      self.pre_process_data()
      self.renew=renew
      peft_config = LoraConfig(
        r=16,
        lora_alpha=32,
        lora_dropout=0.1,
        bias="none",
        task_type=TaskType.CAUSAL_LM,
      )
      bnb_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_use_double_quant=True,
        bnb_4bit_quant_type="nf4",
        bnb_4bit_compute_dtype=torch.bfloat16,
      )


      if renew:
        self.model_id = model_info

        self.tokenizer = AutoTokenizer.from_pretrained(self.model_id)
        self.tokenizer.pad_token = self.tokenizer.eos_token

        self.backbone_model=AutoModelForCausalLM.from_pretrained(
                      self.model_id,
                      torch_dtype=torch.float16,  
                      device_map="auto"
                  )
        # Tải mô hình
        self.model = AutoModelForCausalLM.from_pretrained(
            model_info,
            quantization_config=bnb_config,
            device_map="auto"
          )
        self.model = prepare_model_for_kbit_training(self.model)
        self.model = get_peft_model(self.model, peft_config)
      elif type(model_info)==str:
        self.model_id = model_info
        self.tokenizer = AutoTokenizer.from_pretrained(self.model_id)
        self.tokenizer.pad_token = self.tokenizer.eos_token
        self.model = AutoModelForCausalLM.from_pretrained(
            model_info,
            quantization_config=bnb_config,
            device_map="auto"
          )
        self.model = prepare_model_for_kbit_training(self.model)
        self.model = get_peft_model(self.model, peft_config)

      elif (tokenizer and model_info):
          self.model_id="testing_unknown"
          self.model =model_info
          self.tokenizer=tokenizer
          self.tokenizer.pad_token = self.tokenizer.eos_token
#-------------------------------------------------------------------developing RAG--------------------------------------BEGIN
  def build_rag_dict(self):
      print("🔍 Đang tạo embedding cho các context...")
      self.rag_dict = []
      self.rag_embedder = SentenceTransformer("mixedbread-ai/mxbai-embed-large-v1")

      # Tokenizer để tính token count
      tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

      # Tách từng câu từ các đoạn context
      tmp = []

      for index,j in enumerate(self.contexts):
        if not (isinstance(j, float) and math.isnan(j)):
          if isinstance(j, float):
            print(f"Float tại vị trí {index} giá trị:{j}")
            tmp.append(str(j))
            continue
          sentences = j.split("/n")
          for i in sentences:
            i = i.strip()
            if i != "":
              tmp.append(i)
        else:
          print(f"NaN tại vị trí {index}")


      # Gộp câu ngắn lại
      merged_sentences = []
      buffer = ""

      for i in range(len(tmp)):
          sentence = tmp[i]
          buffer = (buffer + " " + sentence).strip() if buffer else sentence
          token_count = len(tokenizer.tokenize(buffer))

          if token_count >= 50 or i == len(tmp) - 1:
              merged_sentences.append(buffer.strip())
              buffer = ""

      # Đọc file RAG cũ nếu có
      rag_path = "/content/drive/MyDrive/HuyAnh/RAG_dict/RAG.xlsx"
      old_rag = []

      if os.path.exists(rag_path):
          print("📄 Đã tìm thấy file RAG cũ, đang gộp dữ liệu...")
          df_old = pd.read_excel(rag_path)
          for _, row in df_old.iterrows():
              old_rag.append({
                  "original_context": row["original_context"],
                  "embedding": eval(row["embedding"]) if isinstance(row["embedding"], str) else row["embedding"]
              })

      # Encode RAG mới
      new_rag = []
      for context in np.unique(merged_sentences):
          if isinstance(context, str) and context.strip() != "":
              embedding = self.rag_embedder.encode(context, convert_to_tensor=False).tolist()
              new_rag.append({
                  "original_context": context,
                  "embedding": embedding
              })

      # Gộp & lọc unique theo original_context
      combined_rag = {item["original_context"]: item for item in (old_rag + new_rag)}
      self.rag_dict = list(combined_rag.values())

      print(f"✅ Đã tạo RAG context cho {len(self.rag_dict)} mục.")



  def retrieve_context_from_question(self,question, top_k=4,number_sent=1):
    question_embedding = self.rag_embedder.encode(question, convert_to_tensor=True)

    scored_contexts = []
    for item in self.rag_dict:
        ctx_emb = torch.tensor(item["embedding"], device=question_embedding.device)
        score = util.pytorch_cos_sim(question_embedding, ctx_emb).item()
        scored_contexts.append((score, item["original_context"]))

    scored_contexts.sort(reverse=True, key=lambda x: x[0])
    top_contexts = [ctx for _, ctx in scored_contexts[:top_k] if len(ctx)>20]
    if len(top_contexts)<number_sent:
      number_sent=len(top_contexts)
    return ".".join(top_contexts[0:number_sent])

  def Re_Format_Excel_rag_dict(self):
    if isinstance(self.rag_dict["embedding"].iloc[0], str):
        self.rag_dict["embedding"] = self.rag_dict["embedding"].apply(ast.literal_eval)
    self.rag_dict = self.rag_dict.to_dict(orient="records")


#-------------------------------------------------------------------developing RAG--------------------------------------END


  def trainer_setup(self
                    ,evaluation_strategy="epoch"
                    ,learning_rate=5e-5
                    ,weight_decay=0.01
                    ,num_train_epochs=2
                    ,per_device_train_batch_size=1
                    ,save_strategy="epoch"
                    ,optim="adamw_torch"):

    self.training_args = TrainingArguments(
        output_dir=self.checkpoint_dir,
        learning_rate=learning_rate,
        weight_decay=weight_decay,
        gradient_accumulation_steps=16,
        per_device_train_batch_size=4,
        num_train_epochs=num_train_epochs,
        save_total_limit=1,
        bf16=True,
        optim="paged_adamw_8bit",  # nếu dùng LoRA Q4
        report_to="none",
        label_names=["labels"]
    )

    # Khởi tạo Trainer
    self.trainer = Trainer(
        model=self.model,
        args=self.training_args,
        train_dataset=self.train_dataset,
        eval_dataset=self.val_dataset
    )

  def fit(self,
          evaluation_strategy="steps"
          ,learning_rate=5e-5
          ,weight_decay=0.01
          ,num_train_epochs=2
          ,per_device_train_batch_size=1
          ,save_strategy="epoch"
          ,optim="adamw_torch"
          ,data_next=None):

    if data_next:
      self.data_list=data_next
      self.pre_process_data()
    if len(self.conservations_list) == 0:
      raise ValueError("❌ Không có dữ liệu hội thoại nào sau khi tiền xử lý!")
    for index,conservations in enumerate(self.conservations_list):
      print(f"\rTiến độ training: {index+1}/{len(self.conservations_list)}\n", end="", flush=True)
      self.model.gradient_checkpointing_enable()  # Giảm bộ nhớ activations trong quá trình training
      self.make_data_train(conservations)
      self.trainer_setup(evaluation_strategy,learning_rate,weight_decay,num_train_epochs,per_device_train_batch_size,save_strategy,optim)
      self.trainer.train()

      self.model = self.trainer.model

      del self.trainer
      gc.collect()
      torch.cuda.empty_cache()


  def make_data_train(self,conversations,test_size=0.1, random_state=42):
    train_data, val_data = train_test_split(conversations, test_size=test_size, random_state=random_state)
    self.train_dataset = ChatDataset(train_data, self.tokenizer)
    self.val_dataset = ChatDataset(val_data, self.tokenizer)

    if len(self.train_dataset) == 0:
      raise ValueError("❌ train_dataset rỗng! Không thể huấn luyện.")
    if len(self.val_dataset) == 0:
      print("⚠️ Cảnh báo: val_dataset rỗng. Training vẫn tiếp tục nhưng không có validation.")
  def cal_token(self,text):
    cnt=0
    for i in text.split(" "):
      cnt+=1
    return cnt

  def pre_process_data(self):
    self.conservations_list=[]
    for data in self.data_list:
      conversations=[]
      for index,ans in enumerate(data["Answer"]):
        if (type(ans) != str) or ans=="" or ans==" ":
          data=data.drop(index)
      self.questions=list(data["Question"])
      self.contexts=list(i for i in data['Context'])
      self.answers=[str(i)+".\n" for i in list(data['Answer'])]
      if self.rag_off:
        self.build_rag_dict()
        self.rag_off=False
      timing=0
      for index,va in enumerate(self.questions):
        if (isinstance(va, float) and math.isnan(va)):
          print(f"Câu hỏi bị NaN tại index{index} với giá trị {va}")
          continue

        start_time = time.time()

        contexts= self.retrieve_context_from_question(va)

        print(f"\rSample thứ {index}/{len(self.questions)} ** {(timing*len(self.questions)-timing*index)/60 if (timing*len(self.questions)-timing*index)>=60 else (timing*len(self.questions)-timing*index)}min",end="", flush=True)

        conversations.append({"context":contexts,"question":va, "answer":self.answers[index]})
        stop_time=time.time()
        timing=stop_time - start_time
      self.conservations_list.append(conversations)

  def save_model(self, save_dir=None):
    if save_dir is not None:
        self.save_directory = save_dir

    os.makedirs(self.save_directory, exist_ok=True)

    has_adapter = hasattr(self.model, "merge_and_unload")
    is_quantized = hasattr(self.model, "quantization_method") or any(
        "4bit" in str(type(m)).lower() or "bnb" in str(type(m)).lower()
        for m in self.model.modules()
    )

    if has_adapter and is_quantized:
        adapter_save_path = "/content/drive/MyDrive/HuyAnh/Model_Finetuning_RENEW/adapter"

        # 1️⃣ Lưu adapter hiện tại
        print(f"💾 Đang lưu adapter vào: {adapter_save_path}")
        os.makedirs(adapter_save_path, exist_ok=True)
        self.model.save_pretrained(adapter_save_path)

        # 2️⃣ Merge adapter vào mô hình backbone FP16
        if self.backbone_model is not None:
            print("🔁 Đang merge adapter vào backbone FP16...")
            from peft import PeftModel

            merged_model = PeftModel.from_pretrained(self.backbone_model, adapter_save_path)
            merged_model = merged_model.merge_and_unload()

            # 3️⃣ Lưu mô hình đã merge
            print(f"💾 Đang lưu mô hình đã merge vào: {self.save_directory}")
            merged_model.save_pretrained(self.save_directory)
            self.tokenizer.save_pretrained(self.save_directory)
            print("✅ Đã lưu mô hình hoàn chỉnh (merge từ QLoRA).")
            return
        else:
            print("❌ Không có backbone_model để merge.")
            return

    # Nếu không phải QLoRA (model không quant hoặc đã merge rồi)
    if has_adapter:
        print("🔁 Đang merge adapter (non-quant) vào mô hình...")
        model_to_save = self.model.merge_and_unload()
    else:
        print("💾 Mô hình không có adapter. Lưu trực tiếp.")
        model_to_save = self.model

    print(f"💾 Đang lưu mô hình vào: {self.save_directory}")
    model_to_save.save_pretrained(self.save_directory)
    self.tokenizer.save_pretrained(self.save_directory)
    print("✅ Đã lưu mô hình hoàn chỉnh.")




  def ask_question(
      self,
      question,
      max_new_tokens=1000,
      num_beams=5,
      temperature=0.9,
      top_k=30,
      no_repeat_ngram_size=3,
      repetition_penalty=1.5,
      early_stopping=True,
      top_p=0.8,
      save_strategy="epoch"
  ):
      # Truy xuất context liên quan
      context = self.retrieve_context_from_question(question)

      # Tạo prompt chuẩn theo LLaMA 3 Instruct
      prompt = (
          "<|begin_of_text|><|start_header_id|>system<|end_header_id|>\n"
          f"{self.system_prompt}<|eot_id|><|start_header_id|>user<|end_header_id|>\n"
          f"{context}\n\n{question}<|eot_id|><|start_header_id|>assistant<|end_header_id|>\n"
      )

      # Tokenize & đẩy lên đúng device
      inputs = self.tokenizer(prompt, return_tensors="pt")
      device = self.model.device  # đảm bảo đồng bộ với model
      inputs = {k: v.to(device) for k, v in inputs.items()}

      # Sinh câu trả lời
      outputs = self.model.generate(
          input_ids=inputs["input_ids"],
          attention_mask=inputs["attention_mask"],
          max_new_tokens=max_new_tokens,
          num_beams=num_beams,
          pad_token_id=self.tokenizer.eos_token_id,
          eos_token_id=self.tokenizer.eos_token_id,
          temperature=temperature,
          do_sample=True,
          top_k=top_k,
          top_p=top_p,
          no_repeat_ngram_size=no_repeat_ngram_size,
          repetition_penalty=repetition_penalty,
          early_stopping=early_stopping
      )

      # Decode kết quả
      decoded_output = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

      return decoded_output

  # ✅ Hàm tính Perplexity
  def calculate_perplexity(self,dataset, model, tokenizer):
      """Tính Perplexity cho từng mẫu trong dataset"""
      model.eval()
      ppl_scores = []

      for idx in range(len(dataset)):
          data_sample = dataset[idx]
          input_ids = data_sample["input_ids"].unsqueeze(0).to("cuda")
          attention_mask = data_sample["attention_mask"].unsqueeze(0).to("cuda")

          with torch.no_grad():
              outputs = model(input_ids, attention_mask=attention_mask, labels=input_ids)
              loss = outputs.loss  # Cross-entropy loss

          ppl = torch.exp(loss).item()
          ppl_scores.append(ppl)

      return sum(ppl_scores) / len(ppl_scores)  # Trả về perplexity trung bình

  # ✅ Hàm tính Factual Accuracy
  def factual_accuracy_score(self,dataset, model, tokenizer):
      """Tính factual accuracy dựa trên độ tương đồng giữa câu trả lời mô hình và câu tham chiếu"""
      accuracy_scores = []

      for idx in range(len(dataset)):
          try:
              question, reference = dataset.get_question_answer(idx)  # Lấy câu hỏi và câu tham chiếu đúng
              if not isinstance(reference, str) or len(reference.strip()) == 0:
                  print(f"❌ Câu tham chiếu lỗi tại index {idx}: {reference}")
                  continue  # Bỏ qua nếu câu tham chiếu lỗi

              # Gọi mô hình để dự đoán câu trả lời
              predicted_response = self.ask_question(question)
              answers = re.findall(r'A: (.*?)(?=Q:|\Z)', predicted_response, re.DOTALL)
              answer = answers[0].strip() if answers else predicted_response.strip()

              if not isinstance(answer, str) or len(answer.strip()) == 0:
                  print(f"❌ Câu trả lời lỗi tại index {idx}: {answer}")
                  continue  # Bỏ qua nếu câu trả lời lỗi

              # Mã hóa câu trả lời mô hình và câu tham chiếu để tính độ tương đồng
              embedding_answer = self.embedder.encode(answer, convert_to_tensor=True)
              embedding_reference = self.embedder.encode(reference, convert_to_tensor=True)

              score = util.pytorch_cos_sim(embedding_answer, embedding_reference).item()
              accuracy_scores.append(score)

          except Exception as e:
              print(f"⚠️ Lỗi tại index {idx}: {e}")

      return sum(accuracy_scores) / len(accuracy_scores) if accuracy_scores else 0  # Tránh chia cho 0

  # ✅ Hàm tính Hallucination Rate
  def hallucination_rate(self,dataset):
      """Tính tỷ lệ hallucination dựa trên Wikipedia"""
      hallucination_scores = []

      for idx in range(len(dataset)):
          question, answer = dataset.get_question_answer(idx)

          # ✅ Kiểm tra nếu dữ liệu bị lỗi (NaN, None, hoặc không phải string)
          if question is None or answer is None or isinstance(question, float) or isinstance(answer, float):
              print(f"❌ Dữ liệu lỗi tại index {idx}: question={question}, answer={answer}")
              continue  # Bỏ qua dữ liệu lỗi

          try:
              # ✅ Xử lý lỗi Wikipedia: nếu không tìm thấy trang, đặt wiki_summary = ""
              wiki_summary = wikipedia.summary(question, sentences=2)
          except wikipedia.exceptions.DisambiguationError as e:
              wiki_summary = wikipedia.summary(e.options[0], sentences=2)
          except wikipedia.exceptions.PageError:
              wiki_summary = ""  # Nếu không có thông tin, đặt rỗng
          except Exception as e:
              print(f"⚠️ Lỗi Wikipedia tại index {idx}: {e}")
              wiki_summary = ""

          # ✅ Kiểm tra nếu dữ liệu lỗi trước khi encode
          if not isinstance(answer, str) or not isinstance(wiki_summary, str):
              print(f"❌ Lỗi dữ liệu không phải string tại index {idx}: answer={answer}, wiki_summary={wiki_summary}")
              continue

          # ✅ Encode câu trả lời và Wikipedia để tính toán độ tương đồng
          embedding_answer = self.embedder.encode(answer, convert_to_tensor=True)
          embedding_wiki = self.embedder.encode(wiki_summary, convert_to_tensor=True)

          similarity_score = util.pytorch_cos_sim(embedding_answer, embedding_wiki).item()
          hallucination_score = 1 - similarity_score  # 1 - độ chính xác

          hallucination_scores.append(hallucination_score)

      # ✅ Tránh lỗi chia cho 0
      if not hallucination_scores:
          print("⚠️ Không có dữ liệu hợp lệ trong dataset. Trả về 0.")
          return 0

      return sum(hallucination_scores) / len(hallucination_scores)
  # ✅ Hàm tính Relevance Score
  def relevance_score(self,dataset):
      """Tính mức độ liên quan giữa câu hỏi và câu trả lời"""
      relevance_scores = []

      for idx in range(len(dataset)):
          question, answer = dataset.get_question_answer(idx)

          # ✅ Kiểm tra nếu question hoặc answer bị lỗi (None, NaN, float)
          if question is None or answer is None or isinstance(question, float) or isinstance(answer, float):
              print(f"❌ Dữ liệu lỗi tại index {idx}: question={question}, answer={answer}")
              continue  # Bỏ qua dữ liệu lỗi

          # ✅ Nếu là float, chuyển thành str
          if isinstance(question, float):
              question = str(question)
          if isinstance(answer, float):
              answer = str(answer)

          # ✅ Encode câu hỏi và câu trả lời
          embedding_question = self.embedder.encode(question, convert_to_tensor=True)
          embedding_answer = self.embedder.encode(answer, convert_to_tensor=True)

          score = util.pytorch_cos_sim(embedding_question, embedding_answer).item()
          relevance_scores.append(score)

      # ✅ Tránh lỗi chia cho 0 nếu danh sách rỗng
      if not relevance_scores:
          print("⚠️ Không có dữ liệu hợp lệ trong dataset. Trả về 0.")
          return 0

      return sum(relevance_scores) / len(relevance_scores)
  def evaluate(self):
    self.embedder=SentenceTransformer("BAAI/bge-large-en-v1.5")
    ppl_score = self.calculate_perplexity(self.val_dataset,self.model,self.tokenizer)
    accuracy_score = self.factual_accuracy_score(self.val_dataset,self.model,self.tokenizer)
    hallucination_score = self.hallucination_rate(self.val_dataset)
    rel_score = self.relevance_score(self.val_dataset)
    model_name=self.model_id.split("/")[-1]

    new_info={
        "Model":model_name,
        "Perplexity Score":ppl_score,
        "accuracy_score":accuracy_score,
        "Hallucination Score":hallucination_score,
        "Relevance Score":rel_score

    }
    if os.path.exists(self.report_path):
      report = pd.read_excel(self.report_path)
      report = pd.concat([report, pd.DataFrame([new_info])], ignore_index=True)
    else:
      report = pd.DataFrame([new_info])
    # Ghi lại vào file Excel (ghi đè)
    report.to_excel(self.report_path, index=False)
    RAG_dict=pd.DataFrame(self.rag_dict)
    RAG_dict['embedding'] = RAG_dict['embedding'].apply(str)
    tmp=len(os.listdir("/content/drive/MyDrive/HuyAnh/RAG_dict"))
    RAG_dict.to_excel(f"/content/drive/MyDrive/HuyAnh/RAG_dict/RAG_{tmp}.xlsx", index=False)

    del self.embedder
    gc.collect()
    torch.cuda.empty_cache()
    print("Đã thêm dòng mới thành công.")

```
